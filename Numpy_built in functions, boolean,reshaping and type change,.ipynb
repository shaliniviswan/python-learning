```python
import numpy as np
a=np.array([[1,2,4],
           [1,7,9],
            [2,7,2]])
a.sum()
a.mean()
a.max()
a.min()
print(a.sum())
print(a.mean())
print(a.max())
print(a.min())
```

```python
import numpy as np
a=np.array([[1,2,4],
           [1,7,9],
            [2,7,2]])
a.sum(axis=0)
a.mean(axis=0)
a.max(axis=0)
a.min(axis=0)
print(a.sum(axis=0))
print(a.mean(axis=0))
print(a.max(axis=0))
print(a.min(axis=0))
```

```python
import numpy as np
a=np.array([[1,2,4],
           [1,7,9],
            [2,7,2]])
print(a[a>2])
```

```python
import numpy as np
a=np.array([[1,2,4],
           [1,7,9],
            [2,7,2]])
a.reshape(1,9)
a.reshape(3,3)
a.reshape(3,-1)
print(a.reshape(1,9))
print(a.reshape(9,1))
print(a.reshape(3,-1))
```
